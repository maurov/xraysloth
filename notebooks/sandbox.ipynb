## Matplotlib plots

```python
import matplotlib.pyplot as plt
from sloth.utils.plotter import Plotter
p = Plotter(nrows=1, ncols=3, title='test', titles=['0 left', '1 cen', '2 right'], figsize=(10,3))
import numpy as np
x = np.linspace(0, 20, 201)
y = np.sin(x)
p.plot(x, y, label='sin', win=0)
p.plot(x, y*x, label='sin*x', win=1)
p.plot(x, y*x**2, label='sin*x^2', win=2)
p.plot(x, -y*x**2/10, label='-sin*x^2/10', win=2, side='right')
p.plot(x, y+0.5, win=0)
p.subplots_adjust(top=0.87)
```
